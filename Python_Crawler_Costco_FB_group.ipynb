```python
soup = BeautifulSoup(browser.page_source,"lxml")
postList = soup.find(class_="j83agx80 l9j0dhe7 k4urcfbm")


js = 'window.scrollTo(0, document.body.scrollHeight)'

    
postList = soup.find_all(class_="j83agx80 l9j0dhe7 k4urcfbm")
postN = len(postList)
print(postN)
while postN < 50:
    browser.execute_script(js)
    time.sleep(1)
    soup = BeautifulSoup(browser.page_source,"lxml")
    postList = soup.find_all(class_="j83agx80 l9j0dhe7 k4urcfbm")
    postN = len(postList)
    print(postN)
    
# for post in postList:
#     print(post.text)
```
